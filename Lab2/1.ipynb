Phase KickBack

```python
import pennylane as qml
from pennylane import numpy as np
dev = qml.device("default.qubit", wires=2)

@qml.qnode(dev)
def phase_kickback(phi):
    # prepare control qubit
    qml.Hadamard(wires=0)
    # prepare eigenstate
    qml.PauliX(wires=1)
    # controlled phase shift
    qml.ctrl(qml.PhaseShift, control=0)(phi, wires=1)
    # interference
    qml.Hadamard(wires=0)
    return qml.probs(wires=0)
```

```python
print(phase_kickback(0))
print(phase_kickback(np.pi/2))
print(phase_kickback(np.pi))
```
